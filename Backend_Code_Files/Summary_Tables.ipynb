```python
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# Dataset and methods
datasets = {
    "SRBCT": [89.62, 90.54, 89.52, 56.02, 80.14],
    "Lymphoma": [91.18, 97.78, 90.91, 90.02, 84.12],
    "Leukemia": [86.99, 95.45, 89.43, 79.06, 69.05],
    "MLL": [89.59, 90.91, 89.43, 79.02, 69.05],
    "Lung": [86.99, 98.14, 91.82, 72.18, 75.01],
    "Ovarian": [91.45, 97.57, 97.37, 86.91, 80.25],
}

methods = ["No FS", "PSO FS", "Filter FS", "GOA FS", "GA FS"]

colors = {
    "No FS": "#f97316",
    "PSO FS": "#60a5fa",
    "Filter FS": "#10b981",
    "GOA FS": "#f472b6",
    "GA FS": "#c084fc",
}

bar_width = 1.2
space_between_datasets = 3  # Increased spacing between dataset groups

# Flatten values
all_vals = []
bar_colors = []
bar_edges = []
x_pos = []

start = 0
for dataset, values in datasets.items():
    for i, val in enumerate(values):
        x_pos.append(start + i)
        all_vals.append(val)
        method = methods[i]
        bar_colors.append(colors[method])
        bar_edges.append("yellow" if method == "PSO FS" else "black")
    start += len(methods) + space_between_datasets

# Plotting
plt.figure(figsize=(22, 8), facecolor="#1f2937")
ax = plt.gca()
ax.set_facecolor("#1f2937")

bars = plt.bar(x_pos, all_vals, color=bar_colors, edgecolor=bar_edges, linewidth=2, width=bar_width)

# Value labels
for bar, val in zip(bars, all_vals):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 1.2, f"{val:.2f}",
             ha='center', va='bottom', fontsize=9, fontweight='bold', color='white')

# X-ticks
xtick_positions = []
xtick_labels = []
start = 0
for dataset in datasets:
    center = start + (len(methods) - 1) / 2
    xtick_positions.append(center)
    xtick_labels.append(dataset)
    start += len(methods) + space_between_datasets

# Dataset separators
for i in range(1, len(datasets)):
    sep = i * (len(methods) + space_between_datasets) - space_between_datasets / 2
    plt.axvline(x=sep, color='white', linestyle='--', linewidth=1)

# Axes and labels
plt.xticks(xtick_positions, xtick_labels, fontsize=11, fontweight='bold', color='white')
plt.yticks(color='white')
plt.ylabel("Accuracy (%)", fontsize=12, color='white')
plt.title("Accuracy Comparison Across Datasets and Feature Selection Methods",
          fontsize=14, fontweight='bold', color='white')
plt.ylim(0, 105)
plt.grid(axis='y', linestyle='--', alpha=0.4, color='white')

# Legend
legend_elements = [
    Patch(facecolor=colors[method], edgecolor='yellow' if method == "PSO FS" else 'black', label=method)
    for method in methods
]

plt.legend(
    handles=legend_elements,
    loc='lower right',
    fontsize=10,
    frameon=True,
    facecolor='white',
    edgecolor='black'
)

plt.tight_layout()
plt.show()
```
